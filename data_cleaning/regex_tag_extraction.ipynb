```python
import re
import pandas as pd
from cleaning import domain
from files import list_to_file, file_to_list
from regex import regex_filter, apply_regexs, generate_regex
```

```python
def category_tags_to_file(terms_list, tag_search_domain, category_file_name):
    regexs = generate_regex(terms_list)
    tags = apply_regexs(regexs, tag_search_domain)
    list_to_file(category_file_name, tags)
    return tags
```

```python
data = pd.read_csv("Data/german-canteens(filtered).csv", sep="@", infer_datetime_format=True, index_col=0)
# evaluates tag_list as list (from string)
data["tags"] = list(map(eval, data["tags"]))

missed_tag_domain = domain(data["tags"]) #file_to_list("Data/tags/tag_domain.txt")
print(len(missed_tag_domain))
```

```python
# set(regex_list(".*vegan.*", nc_tag_domain))-set(regex_list(".*vegetarisch.*", nc_tag_domain))
terms_vgn = ["veganuary", ("veganes", "menü"), ("vegan", "menü")]


# using category_tags_to_file(terms_vgn, missed_tag_domain, "additional_vgn_tags.txt") whould have same effect as
vgn_regexs = generate_regex(terms_vgn)
vgn = apply_regexs(vgn_regexs, missed_tag_domain)
vgn = list(set(vgn)-set(regex_filter(".*vegetarisch.*|.*vegetarian.*", missed_tag_domain)))
print(vgn)
list_to_file("Data/tags/additional_vgn_tags.txt", vgn)

missed_tag_domain_new = list(set(missed_tag_domain)-set(vgn))
```
